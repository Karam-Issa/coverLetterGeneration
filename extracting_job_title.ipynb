```python
import spacy
import pandas as pd

def extract_job_title(text):
    nlp = spacy.load("en_core_web_sm")
    doc = nlp(text)

    # Define patterns for identifying job titles
    title_patterns = [
        [{"POS": "NOUN", "OP": "+"}, {"POS": "PROPN", "OP": "+"}],
        [{"POS": "PROPN", "OP": "+"}, {"POS": "NOUN", "OP": "+"}]
    ]

    # Extract job titles based on patterns
    matcher = spacy.matcher.Matcher(nlp.vocab)
    for i, pattern in enumerate(title_patterns):
        matcher.add(f"JobTitle_{i}", [pattern])

    matches = matcher(doc)
    if matches:
        start, end = matches[0][1], matches[0][2]
        return doc[start:end].text

    return None
```

```python
import spacy
from spacy.matcher import Matcher
import pandas as pd

nlp = spacy.load("en_core_web_sm")

def extract_job_title(text):
    # Define domain-specific keywords
    data_science_keywords = ['data', 'science', 'scientist', 'analyst', 'engineer', 'machine']

    # Define patterns based on common structures and domain-specific keywords
    pattern = [
        {"POS": {"in": ["NOUN", "PROPN"]}, "OP": "+"},
        {"POS": {"in": ["ADJ", "NOUN", "PROPN"]}, "OP": "*"},
        {"POS": {"in": ["NOUN", "PROPN"]}, "OP": "+"},
        {"LOWER": {"in": data_science_keywords}}
    ]

    matcher = Matcher(nlp.vocab)
    matcher.add("JobTitle", [pattern])

    doc = nlp(text)
    matches = matcher(doc)

    # Extract matched phrases and filter based on relevance and word count
    job_titles = [doc[start:end].text for _, start, end in matches if doc[start:end].text.lower() not in ["job", "role", "description", "responsibilities"]]

    # Filter titles with more than 3 words
    job_titles = [title for title in job_titles if len(title.split()) <= 3]

    return job_titles
```

```python
keywords_df = pd.DataFrame()
keywords_df['Job_Title_Keywords'] = df['Preprocessed_Description'].apply(extract_job_title)
```

```python
import pandas as pd
from collections import Counter

# Function to get the top job titles based on frequency
def get_top_job_titles(job_titles, top_n=5):
    # Flatten the list of job titles
    flat_job_titles = [title for sublist in job_titles for title in sublist]

    # Count the occurrences of each job title
    title_counts = Counter(flat_job_titles)

    # Get the top N job titles
    top_titles = title_counts.most_common(top_n)

    return top_titles

# Apply the function to your DataFrame
keywords_df['Top_Job_Titles'] = keywords_df['Job_Title_Keywords'].apply(lambda x: get_top_job_titles(x, top_n=3))

# Expand the 'Top_Job_Titles' column into separate columns
top_job_titles_df = pd.DataFrame(keywords_df['Top_Job_Titles'].tolist(), columns=['Top_Job_Title_1', 'Top_Job_Title_2', 'Top_Job_Title_3'])

# Concatenate the new DataFrame with the original DataFrame
result_df = pd.concat([keywords_df, top_job_titles_df], axis=1)

# Drop the intermediate 'Top_Job_Titles' column if desired
result_df = result_df.drop(columns=['Top_Job_Titles'])
```
